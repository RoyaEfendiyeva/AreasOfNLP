### Machine Translation

```python
from deep_translator import GoogleTranslator

# List of supported languages
SUPPORTED_LANGUAGES = {
    'en': 'İngilis',
    'de': 'Alman',
    'az': 'Azərbaycan',
    'fr': 'Fransız',
    'es': 'İspan',
    'ru': 'Rus',  
}

def translate_text(text, source_lang="en", target_lang="de"):
    try:
        translator = GoogleTranslator(source=source_lang, target=target_lang)
        translated_text = translator.translate(text)
        return translated_text
    except Exception as e:
        print(f"An error occurred while translating: {e}")
        return None

def get_language_code(prompt):
    """Prompt the user to select a language code"""
    while True:
        print("\nSupported languages:")
        for code, lang in SUPPORTED_LANGUAGES.items():
            print(f"{code}: {lang}")
        
        lang_code = input(prompt).strip()
        if lang_code in SUPPORTED_LANGUAGES:
            return lang_code
        else:
            print("Unsupported language code. Please try again.")

def main():
    source_lang = get_language_code("Select the source language (eg 'en' for English): ")
    target_lang = get_language_code("Select the target language (eg 'ru' for Russian): ")

    while True:
        text = input("\nText to be translated (type 'exit' to exit): ").strip()
        if text.lower() == 'exit':
            print("Exiting the program.")
            break
        
        translated_text = translate_text(text, source_lang, target_lang)
        
        if translated_text:
            print("\nOriginal Text:", text)
            print("\nTranslated Text:", translated_text)
        else:
            print("Translation failed. Please try again.")
try :
    if __name__ == "__main__":
         main()
except KeyboardInterrupt:
    print("the program is stopping")
```
