```python
import re

# A dictionary with keywords in list format
responses = [
    (['hi', 'hello', 'hey', 'what\'s up', 'howdy'], 'Hello! How can I assist you today?:)'),
    (['name', 'who are you', 'what is your name'], 'I am a customer support chatbot here to help you. My name is RBot :)'),
    (['how are you', 'how are you doing', 'what\'s up'], 'I am here and ready to help you with any issues you might have! :)'),
    (['can you help me', 'i need help', 'assist me'], 'Absolutely! Please let me know what you need assistance with.:)'),
    (['what is your return policy', 'return policy', 'how do I return an item'], 'Our return policy allows you to return items within 30 days of purchase. For more details, please visit our return policy page.:)'),
    (['how can i contact customer service', 'contact customer service', 'customer service'], 'You can contact our customer service team via email at support@example.com or by calling 1-800-123-4567.:)'),
    (['what are your store hours', 'store hours', 'when is your store open'], 'Our store hours are Monday to Friday, from 9 AM to 6 PM. We are closed on weekends and public holidays.:)'),
    (['how can i update my account information', 'update account', 'change my account details'], 'To update your account information, please log in to your account and visit the "Account Settings" page.:)'),
    (['do you offer international shipping', 'international shipping', 'ship to other countries'], 'Yes, we offer international shipping to most countries. Please check our shipping policy for more details.:)'),
    (['where is my refund', 'refund status', 'when will i get my refund'], 'Refunds are typically processed within 7-10 business days. If it has been longer, please contact us with your order details so we can assist you.:)'),
    (['quit', 'exit', 'goodbye', 'see you later'], 'Goodbye! If you need further assistance, feel free to reach out again. Have a great day! :)'),
    
    # New questions
    (['what payment methods do you accept', 'payment methods', 'how can i pay'], 'We accept various payment methods including credit cards, debit cards, and PayPal. For more details, please visit our payment options page.:)'),
    (['can i change my order', 'modify order', 'change order'], 'Once an order is placed, it cannot be modified. However, you can contact customer service for assistance with cancellations or changes.:)'),
    (['how can i reset my password', 'reset password', 'forgot password'], 'To reset your password, go to the "Forgot Password" page on our website and follow the instructions.:)'),
    (['do you have any discounts', 'discounts', 'promotions'], 'We occasionally offer discounts and promotions. Please check our website or subscribe to our newsletter for the latest updates.:)'),
    (['how can i return an item', 'return item', 'return process'], 'To return an item, please visit our returns page and follow the instructions provided there.:)'),
    (['what should i do if I received a damaged item', 'damaged item', 'received damaged item'], 'If you received a damaged item, please contact customer service immediately with your order number and a description of the damage.:)'),
    (['can i get a replacement item', 'replacement item', 'request replacement'], 'Yes, if you received a damaged or incorrect item, you can request a replacement by contacting our customer service team.:)'),
    (['how long does shipping take', 'shipping time', 'delivery time'], 'Shipping times vary depending on your location. Typically, domestic orders take 3-5 business days to arrive, while international orders may take longer.:)'),
    (['where is your store located', 'store location', 'find store'], 'Our physical store is located at 123 Main Street, City, Country. For store hours and directions, please visit our store locator page.:)'),
    (['do you offer gift cards', 'gift cards', 'buy gift cards'], 'Yes, we offer gift cards in various denominations. You can purchase them on our website.:)'),
    (['can i track my order status', 'track order', 'order status'], 'Yes, you can track your order status by entering your order number on our tracking page.:)'),
    (['how do i unsubscribe from emails', 'unsubscribe', 'email unsubscribe'], 'To unsubscribe from our emails, click the "unsubscribe" link at the bottom of any email weâ€™ve sent you.:)'),
    (['what should i do if i forgot my username', 'forgot username', 'username recovery'], 'If you forgot your username, please use the "Forgot Username" feature on the login page or contact customer service for assistance.:)')
]

def clean_input(user_input):
    # Regex to clean characters and spaces
    cleaned_input = re.sub(r'\s+', ' ', re.sub(r'[^\w\s]', '', user_input)).strip().lower()
    return cleaned_input

def chat():
    print("Hello! I'm a chatbot. Type 'quit' to exit.")
    while True:
        try:
            user_input = input("You: ")
            if user_input.lower() == "quit":
                print("Chatbot: Goodbye! If you need further assistance, feel free to reach out again. Have a great day! :)")
                break
            
            cleaned_input = clean_input(user_input)
            response = "Sorry, I do not understand what you are saying."
            
            # We find the answer with suitable question options
            for patterns, response_text in responses:
                # We check keywords with a nested period
                for pattern in patterns:
                    if pattern in cleaned_input:
                        response = response_text
                        break
                if response != "Sorry, I do not understand what you are saying.":
                    break
            
            print("User:", cleaned_input)
            print("Chatbot:", response)
        
        except KeyboardInterrupt:
            print("\nChatbot: Goodbye! If you need further assistance, feel free to reach out again. Have a great day! :)")
            break

# Start the chatbot
chat()
```
